```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Create a reference the CSV file desired
csv_path = Path("Resources/JC-202305-citibike-tripdata.csv")

# Read the CSV into a Pandas DataFrame
May2023_df = pd.read_csv(csv_path, low_memory=False)

# Print data to the screen
May2023_df
```

```python
# Check Count

May2023_df.count()
```

```python
# clean May2023_df & Check Count

May2023_cleaned_df = May2023_df.dropna(how="any")
May2023_cleaned_df.count()
```

```python
# Export file as .csv

May2023_cleaned_df.to_csv("Resources/May2023_citibike_cleaned.csv", index=False, header=True)
```
